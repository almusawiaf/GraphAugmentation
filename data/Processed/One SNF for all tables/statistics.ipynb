```python
import pandas as pd

df = pd.read_csv('Lung/clusters.csv', index_col=0)


# Count the occurrences of 0, 1, and 2 in each cell
count = df.apply(pd.Series.value_counts).fillna(0)

# Calculate the percentage of occurrences
percentage = (count / count.sum()) * 100

# Display the result
print(percentage)
```
